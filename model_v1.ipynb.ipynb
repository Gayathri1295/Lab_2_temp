```python
#importing funtions
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import classification_report
```

```python
#importing dataset
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset.head()
```

```python
#defining features and  targeted Values
X = dataset.iloc[:, [2, 3]].values
Y = dataset.iloc[:,-1].values
```

```python
#spliting of dataset in test and train
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 25)
```

```python
#feature scalling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
#fitting logistic regression to the Training Set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 25)
classifier.fit(X_train, Y_train)
```

```python
#prediciting the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
#making the Confusion Matrix
from sklearn.metrics import confusion_matrix
from matplotlib.colors import LinearSegmentedColormap

# Calculate confusion matrix
cm = confusion_matrix(Y_test, y_pred)

# Display confusion matrix with a pink color
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='pink', fmt='g', cbar=False,
            xticklabels=['Predicted Negative', 'Predicted Positive'],
            yticklabels=['Actual Negative', 'Actual Positive'])
plt.xlabel('Predicted labels')
plt.ylabel('Actual labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Generate classification report
report = classification_report(Y_test, y_pred)

# Print classification report
print("Classification Report:")
print(report)
```
